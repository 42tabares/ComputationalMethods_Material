```python
import pandas as pd
import numpy as np
import sympy as sy
```

```python
st=pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ6ltUxH161xsPuP_-66A86omeh9xce1JFD0Ok2mhO_JdwrPI4bMwlJPWgpxKsS6-doAS-0nhmLshJg/pub?output=csv')
```

## Convert to sympy objects

```python
for x in ['x1','x2','x3','x4','x5']:
    st[x]=st[x].map(sy.sympify)
```

### Find solutions with two repeated charges

```python
xs=['x1','x2','x3','x4','x5']
st2=pd.DataFrame()
for x in xs:
    for y in xs:
        if x!=y:
            st['pair']=x+'='+y
            tmp=st[ st[x]==st[y] ].reset_index(drop=True)
            tmp['nu']=tmp[x]
            st2=st2.append(tmp).reset_index(drop=True)
            
st2=st2.drop_duplicates(xs).reset_index(drop=True)      
```

```python
st2eq=pd.DataFrame()
for x in ['x1','x2','x3','x4','x5']:
    for y in ['x1','x2','x3','x4','x5']:
        if x!=y:
            st2['equal']=x+'='+y
            st2eq=st2eq.append(st2[ st2[x]==st2[y] ]).reset_index(drop=True)
```
